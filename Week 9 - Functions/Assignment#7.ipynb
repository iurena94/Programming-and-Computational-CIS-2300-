```python
#Question 1
def main():  
    #Start by showing balance and asking for the action
    print(f"Your balance is ${balance}")
    path = input("\nEnter 'D' to deposit, 'W' to withdraw, or 'Q' to exit: ").upper()
    
    #Repeat until user exits
    while path != 'Q':
    #testing for user choice and invalid input
        #Choosing to deposit
        if path == 'D':
            depositAmount()
            return main()
        #Choosing to withdraw
        elif path == 'W':
            withdrawAmount()
            return main()
        #Testing for invalid input
        else:
            print("\nError message: Invalid input! \nPlease try again.")
            return main()
    print("\n- - - - - Exiting - - - - - ")
    print('\nThank you for banking with us.')
    
#Withdraw function
def withdrawAmount():
    print("\n - - - Withdrawing - - - ")
    #amount user want to withdraw
    withdraw = float(input('Enter amount here: $'))
    global balance 
    #testing if withdraw is possible
    test = balance - withdraw
    
    if test < 0: #withdrawal is not possible
        print('\nError message: Insufficient balance! \nNo withdrawals made.')
        
    elif test >= 0: #withdrawal is possible
        #subtract from the balance
        balance -= withdraw
        print(' - - Withdrawal Complete - - ')
    
#Deposit function
def depositAmount():
    print('\n - - - - Depositing - - - - ')
    #amount user wants to deposit
    deposit = float(input("Enter ammount here: $"))
    global balance 
    #add amount to the balance
    balance += deposit
    print(' - - - Deposit Complete - - - ')
    
#start of the variables 
print("How may we assist you today?")
balance = 1000 
main()
```

```python
#Question 2
def main():
    #user input
    n, r, order, repeat = userInput()
    #calling function with inputs
    fractional(n,r,order,repeat)

#defining user inputs
def userInput():
    n = int(input("How many different numbers are possible?: "))
    r = int(input("How many numbers are used?: "))
    order = input("Is the order of the numbers important?(Y or N): ").upper()
    repeat = input("Can you repeat a number?(Y or N): ").upper()
    return n, r, order, repeat

#fractional function
def fractional(n,r,order,repeat):
    #defining variables
    bot = 1
    bot2 = 1
    top = 1

    #matching input to the formula
    #formula n! / (n-r)!
    if order == "Y" and repeat == "N":
        #Top factorial
        for x in range(1,n+1):
            top *= x 
        #Bottom factorial
        for x in range(1,(n-r)+1):
            bot *= x
        perm = top/bot
    
    #formula (r+n-1)! / r!(n-1)!
    elif order == "N" and repeat == "Y":
        #Top factorial
        for x in range(1,(r+n)):
            top *= x
        #First bottom factorial
        for x in range(1,r+1):
            bot *= x
        #Second bottom factorial
        for x in range(1,n):
            bot2 *= x
        #result
        perm = top/(bot*bot2)
    
    #formula n! / r!(n-r)!
    elif order == "N" and repeat == "N":
        #Top factorial
        for x in range(1,n+1):
            top *= x
        #First bottom factorial
        for x in range(1,r+1):
            bot *= x
        #Second bottom factorial
        for x in range(1,(n-r)+1):
            bot2 *= x
        #result
        perm = top/(bot*bot2)
    
    #formula n^r
    elif order == "Y" and repeat == "Y":
        perm = n**r
    return print(f"There are {perm} permutations")

main()
```

```python
## Question 3
def main():
    x = getInput()
    odd(x)
    even(x)
    square(x)

def getInput():
    x = int(input("Enter a number here: "))
    print('Number:',x)
    return x
def odd(x):
    print("Odd:", x % 2 != 0)
    
def even(x):
    print('Even:', x % 2 == 0)
def square(x):
    print('Squared:', x**2)

main()
```

```python
#Question 4
def main():
    #getting the range's start and end from function
    start, end = getInputs()
    #calculating sum in function
    addNumsInRange(start, end)
    
#Sum of range from user input
def addNumsInRange(start, end):
    sum = 0 #defining sum
    for i in range(start, end+1): #going from range 'start' to 'end'
        sum += i
    return print('The sum of the range is:',sum)

#defining the start and end of the range
def getInputs():
    start = int(input("The range starts at: ")) #start variable
    end = int(input("And ends at: ")) #end variable
    return start, end

main()
```
