```python
#Question 1 - Combinations and Permutations
#getting inputs 
n = int(input("How many different numbers are possible?: "))
r = int(input("how many numbers are used?: "))
order = input("Is the order of the numbers important?(Y or N): ").upper()
repeat = input("Can you repeat a number?(Y or N): ").upper()

#defining variables
bot = 1
bot2 = 1
top = 1

#matching input to the formula
#formula n! / (n-r)!
if order == "Y" and repeat == "N":
    
    #Top factorial
    for x in range(1,n+1):
        top *= x 
    
    #Bottom factorial
    for x in range(1,(n-r)+1):
        bot *= x
        
    perm = top/bot
    
#formula (r+n-1)! / r!(n-1)!
elif order == "N" and repeat == "Y":
    
    #Top factorial
    for x in range(1,(r+n)):
        top *= x
    
    #First bottom factorial
    for x in range(1,r+1):
        bot *= x
    
    #Second bottom factorial
    for x in range(1,n):
        bot2 *= x

    perm = top/(bot*bot2)
    
#formula n! / r!(n-r)!
elif order == "N" and repeat == "N":
    
    #Top factorial
    for x in range(1,n+1):
        top *= x
    
    #First bottom factorial
    for x in range(1,r+1):
        bot *= x
    
    #Second bottom factorial
    for x in range(1,(n-r)+1):
        bot2 *= x
    
    perm = top/(bot*bot2)
    
#formula n^r
elif order == "Y" and repeat == "Y":
    perm = n**r
    
print(f"There are {perm} permutations")
```

```python
#Question 2 - Pennies for pay
#getting inputs
num_days = int(input('Enter the number of days here: '))
pennies = 0
print('\nDay\tPennies')
print('------|-------------')

#calculating
if num_days >= 1:
    pennies += 1
    print(f'1\t{pennies}')
    for x in range(2,num_days+1):
        pennies *= 2
        print('------|-------------')
        print(f'{x}\t{pennies}')
else:
    print('0\t0')
    
print('====================')
dollars = pennies/100
print('Total = ${:.2f}'.format(dollars))

if dollars >= 5000:
    print(f'\nIt is worth earning pay at this rate if you work for at least {num_days} days because you just made ${dollars}')
```
